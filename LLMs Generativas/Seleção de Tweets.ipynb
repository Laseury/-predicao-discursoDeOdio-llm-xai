## **Importando biblioteca**

```python
# Auxiliaries
import pandas as pd
import random
import time
import datetime
import numpy as np
import io
import re
```

## **Lendo Dataset**

```python
url = 'https://drive.google.com/file/d/1MZwoTp8TQMANfhjphRr65ue5mdpa-nJk/view?usp=sharing'
file_id = url.split('/')[-2]
read_url='https://drive.google.com/uc?id=' + file_id

# read the data
dataset = pd.read_csv(read_url,  index_col=0)

# display the first 5 rows
dataset.head()
```

## **Seleção dos tweets**

```python
# 0 - hate speech
# 1 - offensive language
# 2 - neither

odio =  dataset.loc[dataset['class'] == 0].sample(n = 50)
linguagem_ofensiva =  dataset.loc[dataset['class'] == 1].sample(n = 50)
nenhum = dataset.loc[dataset['class'] == 2].sample(n = 50)

amostras =  list(odio['tweet'].values) + list(linguagem_ofensiva['tweet'].values) + list(nenhum['tweet'].values)
classes = [0]*50 + [1]*50 + [2]*50
```

```python
dic = {'tweet': amostras, 'class': classes}
df = pd.DataFrame(dic)
df.to_excel('amostras.xlsx')
```
